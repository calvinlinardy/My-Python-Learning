```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('fingerprint.jpeg',0)
kernel = np.ones((3,3), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1) 
img_opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
img_closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(dpi=100,figsize=(6,6))
plt.imshow(img,cmap='gray')
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,16))

axes[0][0].imshow(img_closing,cmap='gray')
axes[0][0].title.set_text('Closing')
axes[1][1].imshow(img_opening,cmap='gray')
axes[1][1].title.set_text('Opening')
axes[0][1].imshow(img_erosion,cmap='gray')
axes[0][1].title.set_text('Erosion')
axes[1][0].imshow(img_dilation,cmap='gray')
axes[1][0].title.set_text('Dilation')

fig.subplots_adjust(
    wspace=0.1, # ini space width nya
    hspace=0.1, # ini space height
);
```
