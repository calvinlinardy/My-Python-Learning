```python
from PIL import Image
```

## Pixel Replication

```python
old_image = Image.open('rose.jpeg')
```

```python
old_image
```

```python
print(old_image.size)
print(old_image.format)
```

```python
factor = 10
```

```python
shrink_size = tuple([int(x/factor) for x in old_image.size])
```

```python
# SHRINK
new_image = old_image.resize(shrink_size,Image.NEAREST)
```

```python
new_image
```

```python
zoom_size = tuple([int(x*factor) for x in old_image.size])
```

```python
# ZOOM
original_size_image = new_image.resize(zoom_size,Image.NEAREST)
```

```python
original_size_image
```
