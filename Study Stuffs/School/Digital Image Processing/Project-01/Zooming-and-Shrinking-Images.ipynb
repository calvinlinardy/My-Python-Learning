```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

## Pixel Replication

```python
img_pil = Image.open("rose.jpeg")
```

```python
img_pil.size
```

```python
# SHRINK BY A FACTOR OF 10
factor = 10
shrink_size = tuple([int(x/factor) for x in img_pil.size])
new_img_pil = img_pil.resize(shrink_size,Image.NEAREST)
```

```python
new_img_pil
```

```python
# ZOOM THE IMAGE BACK TO THE ORIGINAL RESOLUTION
original_size_image = new_img_pil.resize(img_pil.size,Image.NEAREST)
original_size_image
```

## Bilinear Interpolation

```python
img = cv2.imread("rose.jpeg")
w,h = img.shape[:2]
```

```python
# SHRINK TO 100 DPI
bil_int = cv2.resize(img,None,fx=(100/w),fy=(100/h),
interpolation=cv2.INTER_LINEAR)

print(f"DPI: {bil_int.shape[:2]}")
plt.figure(dpi=100)
plt.imshow(bil_int)
```

```python
w
```

```python
h
```

```python
# ZOOM BACK TO 1024 DPI
bil_int_zoom = cv2.resize(bil_int,None,fx=w/100,fy=h/100,
interpolation=cv2.INTER_LINEAR)

print(f"DPI: {bil_int_zoom.shape[:2]}")
plt.figure(dpi=100,figsize=(16,16))
plt.imshow(bil_int_zoom)
```
