## Seaborn Parameters

The hue and palette parameters are commonly available around many plot calls in seaborn.

### hue

Color points based off a categorical feature in the DataFrame

jadi hue ini bakalan warnain points nya berdasarkan value dari column yang kita pilih, dan otomatis munculin legend

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education');
# contoh ini kita mau bedain warnanya berdasarkan value dari 
# level of education
```

```python
# skrg kita coba ganti hue nya ke salary
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary');
```

warnanya bakalan jadi gradient karena ini continous value, buka categorical value. jadi continuous kan angka yang dari kecil-besar, kalo kategori kan semua beda2 berdasarkan jenis.

## Scatterplot  Parameters

These parameters are more specific to the scatterplot() call

### size

Allows you to size based on another column 

jadi size ini kayak hue juga, nanti di kategorikan berdasarkan size

ini biasanya buat continous value, semakin gede valuenya, semakin gede sizenya

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience');
```

kalo mau beneran ganti size pointnya secara merata, pake s=

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200);
```

tapi kalo kayak gini pasti ada yg ketumpuk2, kita ganti alpha nya biar keliatan mana yg ketumpuk. alpha = 0 itu ilang, alpha = 1 itu normal kaya skrg, kita kasih alpha antara itu

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2);
# linewidth 0 biar buletnya ga ada bordernya
```

### style

Automatically choose styles based on another categorical feature in the dataset. Optionally use the **markers=** parameter to pass a list of marker choices based off matplotlib, for example: ['*','+','o']`

kalo style di kategorikan berdasarkan bentuk pointnya

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=100,style='level of education');
```

```python
# bisa di kategoriin pake warna juga pake hue
# jadi bisa beberapa sekaligus
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=100,
style='level of education',hue='level of education');
```

```python
# bisa ganti warna juga
plt.figure(figsize=(12,8))
sns.scatterplot(x='salary',y='sales',data=df,s=100,
style='level of education',hue='level of education',palette='Paired');
```

## Exporting

```python
# cara exportnya sama kaya figure matplotlib
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='salary',palette='viridis');

# bebas mau jpg ato png
plt.savefig('example_scatter.jpg')
```
