# Basic Example

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
# kita buat plot nya buat liat relationship x sama y
```

```python
plt.plot(x,y)
# peratiin kalo dibawah ada info tentang memory nya
# ada 2 cara buat ilangin infonya
# 1. pake plt.show() di akhir, sama kaya kalo pake IDE
# 2. pake semicolon di akhir ';'
```

## PARAMETERNYA ADA BANYAK BANNGET, INI BEBERAPA YG SERING DI PAKE

## PENTING!!
### semua line codes nya harus di 1 cell, matplotlip gabisa gabung2 beberapa cell!!!

```python
plt.plot(x,y)
plt.xlabel('X Axis') # buat tambahin title x
plt.ylabel('Y Axis') # buat tambahin title y
plt.title('XY Relationship') # buat tambahin title plot
plt.show() # buat ilangin info RAM
```

### More figure parameters

```python
plt.plot(x,y)
plt.xlabel('X Axis') # buat tambahin title x
plt.ylabel('Y Axis') # buat tambahin title y
plt.title('XY Relationship') # buat tambahin title plot

# ini buat kalo kita mau zoom in plotnya liat di range tertentu

plt.xlim(0,5) # buat kasih limit di x axis dari mana sampe mana
plt.ylim(0,8); # kasih limit y
# kita coba pake semicolon ';'
```
