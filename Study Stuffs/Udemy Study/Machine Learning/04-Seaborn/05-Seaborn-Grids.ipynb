## Pairgrid

Grid that pairplot is built on top of, allows for heavy customization of the pairplot seen earlier.

jadi pairgrid itu kayak pairplot cuma bisa di customize

```python
sns.pairplot(df)
```

```python
sns.PairGrid(df)
# kosong nungguin mapping command
```

```python
# mapping harus make sense

g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
```

```python
g = sns.PairGrid(df)
g = g.map_lower(sns.kdeplot)
```

```python
# map_diag
# diagonal itu valuenya kan garis lurus, jadi kalo scatter pasti ga make sense

g = sns.PairGrid(df)
g = g.map_diag(sns.scatterplot) # makanya ga muncul

# harusnya pake histogram ato scatterplot baru make sense
```

```python
g = sns.PairGrid(df)
g = g.map_diag(sns.histplot)
```

```python
# skrg gabungin
# bisa dikasih parameter lagi buat plotnya
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot, lw=2)
g = g.map_lower(sns.kdeplot, colors="red")
```

```python
# pairgridnya juga bisa dikasih parameter
# dan legend nya musti add manual, kalo ga bingung nanti yg mana warna apa

g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

NOTE SEBENERNYA PAIRGRID GA GITU KEPAKE, INI BIAR TAU AJA KALO ADA FITURNYA, pairplot udh more than enough

## FacetGrid

facetgrid itu kerangka dari catplotnya, kita bisa liat disini

```python
sns.FacetGrid(data=df,col='gender',row='lunch')
```

jadi facetgrid itu bener2 ga terlalu kepake, catplot aja udah lebih dari cukup. jadi pake facetgrid kalo bener2 kita mau fully customize ato sesuati ga working sama apa yang kita mau. facetgrid makan effort yang lebih buat sesuatu yang sama

```python
g = sns.FacetGrid(data=df,col='gender',row='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()
```

```python
g = sns.FacetGrid(data=df,col='gender',row='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()

plt.subplots_adjust(hspace=0.4, wspace=1)
```
