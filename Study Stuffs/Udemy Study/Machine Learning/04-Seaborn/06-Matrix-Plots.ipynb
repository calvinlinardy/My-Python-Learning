## Heatmap

inget heatmap harus bandingin category yang unit nya sama, misal kalo meter jangan di bandingin sama cm, atau misal kaya df ini, rate(persentase) gabisa di bandingin sama life expectancy(umur) nanti jadi kacau range nya

```python
# kita set indexnya jadi countries
df = df.set_index('Countries')
```

```python
df
```

```python
sns.heatmap(df)

# life expectancy beda unit, jadi warna nya absurd sendiri
```

```python
# jadi kita drop
sns.heatmap(df.drop('Life expectancy',axis=1))
```

```python
# tambahin garis biar lebih jelas
sns.heatmap(df.drop('Life expectancy',axis=1),linewidths=0.5)
```

```python
# kita tambahin annot sama gedein dpi
plt.figure(figsize=(10,8),dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True)
```

```python
# buat ganti color palette musti pilih warna yg bener, biar color blind bisa liat
# pake cmap(color mapping), bukan palette
plt.figure(figsize=(10,8),dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True,cmap='viridis')
```

```python
# color bar nya juga bisa di set warnanya
# tapi ga recommend, cuma buat tau aja
sns.heatmap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True,
cmap='viridis',center=40)

# ini maksudnya taro center dari color palette nya di angka 40
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True,
cmap='viridis',center=1)

# taro centermya di 1, jadi yang tadi itu first half, ini second half palette nya
```

## Clustermap

Plot a matrix dataset as a hierarchically-clustered heatmap.

```python
# ngecluster di cari yang similar valuenya
```

```python
sns.clustermap(df.drop('Life expectancy',axis=1))

# ga make sense kalo cluster column nya juga, harusnya cuma cluster indexnya
```

```python
# kita bagusin + ilangin cluster columnya
sns.clustermap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True,
cmap='viridis',col_cluster=False)
```

```python
# bisa ubah position color barnya juga pake cbar_pos
sns.clustermap(df.drop('Life expectancy',axis=1),linewidths=0.5,annot=True,
cmap='viridis',col_cluster=False,cbar_pos=(-0.1,.2,.03,.4))
```
