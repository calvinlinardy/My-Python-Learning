# Comparison Plots with pairplot() and jointplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/clcx/Documents/GitHub/My-Python-Learning/Udemy Repo/Machine Learning/05-Seaborn/StudentsPerformance.csv')
```

## jointplot

```python
# kita butuh continuous features
# kita pake scores nya
```

```python
sns.jointplot(data=df,x='math score',y='reading score')
```

```python
# karena bertumpukan kita bisa pake kind=hex biar lebih jelas
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')
```

```python
# atau kalo tetep mau pake scatter juga bisa kecilin alphanya
sns.jointplot(data=df,x='math score',y='reading score',kind='scatter',
alpha=0.2)

# tapi menurut gw emang agak jelek
```

```python
# atau kalo gamau hex, maunya kotak2 bisa pake hist
sns.jointplot(data=df,x='math score',y='reading score',kind='hist')

# ini lebih bagus soalnya align sama bars nya
```

```python
# bisa juga kind=kde
sns.jointplot(data=df,x='math score',y='reading score',kind='kde')
```

```python
# bisa di shade juga pake color palette
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',shade=True)
```

```python
# kalo kita pake hue dia otomatis barsnya jadi kde, terus ada legend

sns.jointplot(data=df,x='math score',y='reading score',hue='gender')
```

## pairplot

ini bakalan makan banyak CPU dan RAM jadi kadang musti kita bedah dulu datanya pake pandas biar cuma tampilih rows ato columns yg kita mau

```python
sns.pairplot(df)
```

```python
# kalo pake hue yang diagonal jadi kde

sns.pairplot(df,hue='gender',palette='viridis')
```

```python
# kalo tetep mau histogram, bisa set manual

sns.pairplot(df,hue='gender',palette='viridis',diag_kind='hist')
```

```python
# buat ilangin data yg duplicate ato mirror, bisa set corner=True

sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```
