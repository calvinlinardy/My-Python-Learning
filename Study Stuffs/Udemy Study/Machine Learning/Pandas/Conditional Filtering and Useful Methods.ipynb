## Multiple Conditions

Recall the steps:

* Get the conditions
* Wrap each condition in parenthesis
* Use the | or & operator, depending if you want an 
    * OR | (either condition is True)
    * AND & (both conditions must be True)
* You can also use the ~ operator as a NOT operation

```python
# kalo condition nya ada 2 pake or atau and
# tapi kalo di pandas, karena ini ngebandingin series jadi gabisa pake 'or' 'and'
# or jadi |
# and jadi %
# not bisa pake ~ atau ! juga
```

```python
df[(df['day'] == 'Sun') & (df['sex'] == 'Male')].head(5)
```

```python
# GINI
df[(df['day'] == 'Sun') & (df['sex'] != 'Male')].head(5)
```

```python
# SAMA AJA KAYA GINI
df[(df['day'] == 'Sun') & ~(df['sex'] == 'Male')].head(5)
```

```python
# Cari yang weekend
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')].head(5)
```

```python
# kita liat contoh diatas buat cari weekend
# secara ga langsung kan statementnya cocokin apakah day nya itu sun,sat
# ada cara lain yg lebih gampang
# kita buat list yg isinya sun,sat terus pake function 'isin' <- is in
```

```python
options = ['Sat','Sun']
df['day'].isin(options)
```

```python
df[df['day'].isin(options)]
```

#### Contoh
kita mau liat quality dari tip nya apakah banyak atau dikit

```python
def tipper(total_bill,tip): # biasain namain parameter kaya column nya biar readable!!!
    if tip/total_bill > 0.25:
        return 'Orang Kaya'
    else:
        return 'Orang Biasa'
```

```python
# EMANG AGAK TRICKY!!
# musti liat pola sama struktur nya. lama2 pasti biasa
```

```python
# kita buat column baru
df['Tipper'] = df[['total_bill','tip']].apply(lambda df: 
tipper(df['total_bill'],df['tip']),axis=1)
```

```python
df[['total_bill','tip']].apply(lambda df: 
tipper(df['total_bill'],df['tip']),axis=1)
```

```python
# kita test
tipper(16.99,1.01)
```

```python
import timeit
```

```python
setup = '''
import numpy as np
import pandas as pd

df = pd.read_csv(r'E:\\GitHub\\My-Python-Learning\\Machine Learning\\03-Pandas\\tips.csv')

def tipper(total_bill,tip):
    if tip/total_bill > 0.25:
        return 'Orang Kaya'
    else:
        return 'Orang Biasa'
'''

stmt_one = '''
df['Tipper'] = df[['total_bill','tip']].apply(lambda df: tipper(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tipper'] = np.vectorize(tipper)(df['total_bill'],df['tip'])
'''
```

```python
# di atas kita cuma buat siapin codenya aja
# setup itu semua code yg dibutuhin termasuk functionnya
# statement one sama two itu line of codes yg mau di bandingin
```

```python
# hitung waktu stmt one (coba run 1000X)
timeit.timeit(setup = setup, stmt = stmt_one, number = 1000)
```

```python
# hitung waktu stmt two (coba run 1000X)
timeit.timeit(setup = setup, stmt = stmt_two, number = 1000)
```
