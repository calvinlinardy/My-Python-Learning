# The pivot() method

The pivot method reshapes data based on column values and reassignment of the index. Keep in mind, it doesn't always make sense to pivot data. In our machine learning lessons, we will see that our data doesn't need to be pivoted. Pivot methods are mainly for data analysis,visualization, and exploration.

```python
help(pd.pivot)
```

----
#### Note: Common Point of Confusion: Students often just randomly pass in index,column, and value choices in an attempt to see the changes. This often just leads to formatting errors. You should first go through this checklist BEFORE running a pivot():

* What question are you trying to answer?
* What would a dataframe that answers the question look like? Does it need a pivot()
* What you want the resulting pivot to look like? Do you need all the original columns?

## The pivot_table() method

Similar to the pivot() method, the pivot_table() can add aggregation functions to a pivot call.

aggregation ada di section groupby (kalo lupa)

kali ini pass in func nya sebagai string, terus sum ini bakalan jumlahin semua int, jadi account number juga di jumlahin padahal ga makesense

```python
pd.pivot_table(df,index='Company',aggfunc='sum')
```

```python
# bisa juga pake groupby
df.groupby('Company').sum()
```

```python
# kalo kita gamau liat account numbernya, tinggal masukin values ke pivotnya
# jadi cuma tampilin yg kita mau
pd.pivot_table(df,index='Company',aggfunc='sum',values=['Licenses','Sale Price'])
```

## Multiindex

```python
# misal kita mau liat sale price setiap
# account manager dari setiap contact nya
```

```python
pd.pivot_table(df,index=['Account Manager','Contact'],values='Sale Price',aggfunc='sum')
```

Columns are optional - they provide an additional way to segment the actual values you care about. The aggregation functions are applied to the values you list.

```python
pd.pivot_table(df,index=['Account Manager','Contact'],columns='Product',values='Sale Price',aggfunc='sum')
# kita tambah column product
```

```python
# kalo mau ada nama tablenya, masukin ke list []
pd.pivot_table(df,index=['Account Manager','Contact'],columns=['Product'],values=['Sale Price'],aggfunc='sum')
```

```python
# kalo mau jadiin NaN jadi 0
# pake fill_value = 0
pd.pivot_table(df,index=['Account Manager','Contact'],columns=['Product'],values=['Sale Price'],aggfunc='sum',fill_value=0)
```

```python
# kita bisa juga aggfunc nya pake np.sum
pd.pivot_table(df,index=['Account Manager','Contact'],columns=['Product'],values=['Sale Price'],aggfunc=[np.sum],fill_value=0)
```

```python
# bisa juga pake 2 aggfunc, sum sama mean misal
pd.pivot_table(df,index=['Account Manager','Contact'],columns=['Product'],values=['Sale Price'],aggfunc=[np.sum,np.mean],fill_value=0)
```

tapi sebenernya kayak gini tabel nya udah agak berantakan, tujuan pivot itu simplify data bukan bikin jadi complicated, jadi gini cara simplify nya

```python
# hilangin mean nya, sum aja
# sama pindahin product jadi index
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum],fill_value=0)
```

```python
# margin = True
# ini buat nambahin total
# jadi di paling bawah ada total
pd.pivot_table(df,index=['Account Manager','Contact','Product'],values=['Sale Price'],aggfunc=[np.sum],fill_value=0,margins=True)
```
