```python
import folium as fol
import pandas as pd
```

```python
df = pd.read_csv('Volcanoes.txt')
```

```python
df.head()
```

```python
df
```

```python
map = fol.Map(location=[38.58, -99.09],zoom_start=6,tiles='OpenStreetMap')
```

```python
lon = list(df['LON'])
lat = list(df['LAT'])
elev = list(df['ELEV'])
```

```python
fg = fol.FeatureGroup(name='Map')

for lt,ln,el in zip(lat,lon,elev):
    fg.add_child(fol.Marker(location=[lt,ln],popup=str(el)+' m',icon=fol.Icon(color='green')))
map.add_child(fg)
```
