# Matplotlib SubPlots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
# ini agak complicated but stick with me
# jadi subplots ini itu tuples terdiri dari figure,axes nya
asdf = plt.subplots(nrows=1,ncols=2)
```

```python
asdf
# bisa kita liat kan isinya itu tuple
```

```python
# jadi kita buat subplots bisa langsung pisahin aja
fig,axes = plt.subplots()
# subplots itu emang di maksudkan buat beberapa figure jadi 1
# jadi kalo ga ada rows ato cols mending buat figure aja
```

```python
# dan kita bisa liat fig sama axesnya
fig
```

```python
type(axes) # type axesnya itu axes
```

```python
# skrg kita tambahin rows sama cols
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig
```

```python
type(axes) # type axes nya jadi np array
```

```python
# kita liat shapenya
axes.shape
# 1 dimensional soalnya emang cuma ada 2 axes, kanan sama kiri
```

```python
axes
```

```python
# skrg tambahin plot
fig,axes = plt.subplots(nrows=1,ncols=2)

# pake index axesnya, inget axesnya itu array
# [0] = kiri, [1] = kanan
axes[0].plot(x,y)
axes[1].plot(a,b);
```

```python
# skrg kita tambahin rows nya jadi 2x2
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b);
```
